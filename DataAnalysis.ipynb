# 2.which types of stock would provide the best investment

### Do a multi linear model to predict thye ivestment 
### keep MktPrice as the target variable

### Multiple linear regression Assumptions
#### 1. Linear relationship: There exists a linear relationship between the independent variable, x, and the dependent variable, y.

#### 2. Independence: The residuals are independent. In particular, there is no correlation between consecutive residuals in time series data.

#### 3. Homoscedasticity: The residuals have constant variance at every level of x.

#### 4. Normality: The residuals of the model are normally distributed.

## Multilinear Regression on Retail Data

```python
# 1.Linear Relationship

pairs(retail_new, pch = 18, col = "red")
```

```python
# 2. Normality of residuals. It is assumed that the residual errors are regularly distributed.

model <- lm(MktPrice ~ TotMktCap+TotalSales17+TotalSales18+CapEmp+Dividend, data = retail_new)
```

```python
hist(residuals(model), col = "red")
```

```python
plot(model,2)
```

```python
# 3. Homogeneity of residuals variance.It is presummated that the residuals' variance will never change(homoscedasticity).

#create fitted value vs residual plot
plot(fitted(model), residuals(model))

#add horizontal line at 0
abline(h = 0, lty = 2)
```

```python
# 4.Independence of residuals error terms.
vol.res = resid(model)
plot (seq(1,length(vol.res),1),vol.res,xlab="Observation Order",ylab="Residual",type="b")
```

```python
summary(model)
```
